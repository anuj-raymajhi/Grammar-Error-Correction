# Function for clipping


```python
df = df.drop(['Unnamed: 0'], axis = 1)
```

```python
df.insert(3, "clipped_target", "")
```

```python
df.head()
```

```python
def clip_text(text):
    MAX_LEN = 32
    clipped_text = []
    for sentence in text.split("."):
        sentence = sentence.strip()
        if len(sentence.split()) > MAX_LEN:
            words = sentence.split()
            clipped_words = words[:MAX_LEN]
            clipped_sentence = ' '.join(clipped_words)
            clipped_text.append(clipped_sentence)
        elif len(sentence) > 0:
            clipped_text.append(sentence)
    return '. '.join(clipped_text)
```

```python
df['clipped_target'] = df['target'].apply(clip_text)
```

```python
from nltk.translate.bleu_score import sentence_bleu
bleu_score_all = []
bleu_score_huggingface = []
bleu_score_1 = []
bleu_score_2 = []
bleu_score_3 = []
bleu_score_4 = []
glue_score = []
rouge_score = []


for i in range(df.shape[0]):

  ground1 = df['target'][i]
  # Example candidate sentence
  candidate = df['predicted_ed'][i]

  bleu_scores = bleu_metric.compute(predictions=[candidate.split(' ')],references=[[ground1.split(' ')]])
  bleu_score_all.append(bleu_scores['precisions'])
  bleu_score_1.append(sentence_bleu([ground1.split(' ')], candidate.split(' '), weights=(1,0,0,0)))
  bleu_score_2.append(sentence_bleu([ground1.split(' ')], candidate.split(' '), weights=(0.5,0.5,0,0)))
  bleu_score_3.append(sentence_bleu([ground1.split(' ')], candidate.split(' '), weights=(0.33,0.33,0.33,0)))
  bleu_score_4.append(sentence_bleu([ground1.split(' ')], candidate.split(' '), weights=(0.25,0.25,0.25,0.25)))
  bleu_score_huggingface.append(bleu_scores["bleu"])


  # google bleu
  glue_scores = glue_metric.compute(predictions=[candidate.split(' ')],references=[[ground1.split(' ')]])
  glue_score.append(glue_scores['google_bleu'])

  # rouge score
  # predictions = ["hello ther"]
  # references = ["hello there"]
  results = rouge.compute(predictions=[ground1],
                          references=[candidate])
  rouge_score.append(results)
  

```
