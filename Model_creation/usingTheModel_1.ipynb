```python
from transformers import T5Tokenizer, T5ForConditionalGeneration, AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained('t5-base')
```

```python
import torch
torch_device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
t5_model = T5ForConditionalGeneration.from_pretrained('Kau-stuv/t5-grammar-error-correction').to(torch_device)
```

```python
def correct_grammar(input_text,num_return_sequences):
  batch = tokenizer([input_text],truncation=True,padding='max_length',max_length=128, return_tensors="pt").to(torch_device)
  translated = t5_model.generate(**batch,max_length=128,num_beams=4, num_return_sequences=num_return_sequences, temperature=1.5)
  tgt_text = tokenizer.batch_decode(translated, skip_special_tokens=True)
  return tgt_text
```

```python
text = 'The teacher explained the lesson clear and answered all of the students questions.He have a lot of experience in his field and is highly respected.She walked to the park with her dog and enjoy the sunshine.'
print(correct_grammar(text, num_return_sequences=4))
```
